```python
import pandas as pd

# Load the dataset from local storage
from google.colab import files
uploaded = files.upload()

# Read the dataset
df = pd.read_csv(next(iter(uploaded)))

# Display basic info & check missing values
df.info()
print(df.isnull().sum())
```

```python
# Fill missing values in 'fuel_type' with the most common value (mode)
df["fuel_type"].fillna(df["fuel_type"].mode()[0], inplace=True)

# Verify missing values are handled
print(df.isnull().sum())
```

```python
# Remove non-numeric characters
df["price"] = df["price"].replace("[$,]", "", regex=True).astype(float)
df["milage"] = df["milage"].replace("[, mi.]", "", regex=True).astype(float)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distributions
plt.figure(figsize=(12, 5))

# Price Distribution
plt.subplot(1, 2, 1)
sns.histplot(df["price"], bins=30, kde=True, color="blue")
plt.title("Price Distribution")

# Milage Distribution
plt.subplot(1, 2, 2)
sns.histplot(df["milage"], bins=30, kde=True, color="green")
plt.title("Milage Distribution")

plt.show()
```

```python
# Function to remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply to price & milage
df = remove_outliers(df, "price")
df = remove_outliers(df, "milage")

# Check new dataset shape
print(df.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical variables
label_encoders = {}
categorical_columns = ["brand", "fuel_type", "transmission"]

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Display first few rows
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize numerical features
scaler = StandardScaler()
df[["milage", "model_year"]] = scaler.fit_transform(df[["milage", "model_year"]])

# Display first few rows after standardization
df.head()
```

```python
 # Extract the first numeric value (engine size in liters) from the 'engine' column
df["engine"] = df["engine"].str.extract(r'(\d+\.?\d*)')

# Convert to float, handling missing values
df["engine"] = pd.to_numeric(df["engine"], errors="coerce")

# Verify the changes
print(df["engine"].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute correlation matrix
correlation_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Function to evaluate models
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    return {"MAE": mae, "RMSE": rmse, "R2 Score": r2}

# Evaluate all models
models = {
    "Linear Regression": linear_model,
    "Random Forest": best_rf,  # Tuned Random Forest
    "XGBoost": best_xgb        # Tuned XGBoost
}

# Store results
results = {name: evaluate_model(m, X_test, y_test) for name, m in models.items()}

# Display evaluation results
import pandas as pd
results_df = pd.DataFrame(results).T
print(results_df)
```
